Answer for Q.1. :-

The primary goal of Object-Oriented Programming (OOP) is to organize and structure code by representing real-world entities as objects that have attributes (data) and behaviors (methods). OOP promotes the concept of modularity, reusability, and maintainability by encapsulating data and methods into objects.

Answer for Q.2. :-

In Python, an object is an instance of a class. It is a concrete entity that represents a specific instance of a class and has its own unique state and behavior. Objects are created based on the definition provided by the class.

Answer for Q.3. :-

A class in Python is a blueprint or template that defines the properties (attributes) and behaviors (methods) that objects of that class will possess. It serves as a blueprint for creating objects. A class acts as a container that encapsulates data and methods related to a specific concept or entity.

Answer for Q.4. :-

Attributes in a class are variables that store data associated with the class or its instances (objects). They represent the state or characteristics of the objects. Methods, on the other hand, are functions defined within a class that perform specific actions or operations on the data (attributes) of the class. They represent the behaviors or actions that objects can perform.

Answer for Q.5. :-

Class variables in Python are variables that are shared among all instances (objects) of a class. They are defined within the class but outside any methods. Class variables have the same value across all instances of the class.

Instance variables, also known as object variables, are variables that are unique to each instance (object) of a class. They are defined within a class's methods and hold data specific to each instance. Each object can have different values for instance variables.

Answer for Q.6. :- 

The purpose of the self parameter in Python class methods is to refer to the instance (object) of the class itself. It allows the methods to access and manipulate the attributes and methods of the specific instance on which the method is called. The self parameter is always the first parameter in a class method definition.

Implementation of the "Book" class:

# Python code

class Book:
    def __init__(self, title, author, isbn, publication_year, available_copies):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.publication_year = publication_year
        self.available_copies = available_copies

    def check_out(self):
        if self.available_copies > 0:
            self.available_copies -= 1

    def return_book(self):
        self.available_copies += 1

    def display_book_info(self):
        print("Title:", self.title)
        print("Author:", self.author)
        print("ISBN:", self.isbn)
        print("Publication Year:", self.publication_year)
        print("Available Copies:", self.available_copies)
        
In this implementation, the "Book" class is defined with the specified attributes (title, author, isbn, publication_year, available_copies). The class includes three methods: check_out(), return_book(), and display_book_info(), which perform the desired functionalities.

Implementation of the "Ticket" class:

# Python code

class Ticket:
    def __init__(self, ticket_id, event_name, event_date, venue, seat_number, price, is_reserved):
        self.ticket_id = ticket_id
        self.event_name = event_name
        self.event_date = event_date
        self.venue = venue
        self.seat_number = seat_number
        self.price = price
        self.is_reserved = is_reserved

    def reserve_ticket(self):
        if not self.is_reserved:
            self.is_reserved = True

    def cancel_reservation(self):
        if self.is_reserved:
            self.is_reserved = False

    def display_ticket_info(self):
        print("Ticket ID:", self.ticket_id)
        print("Event Name:", self.event_name)
        print("Event Date:", self.event_date)
        print("Venue:", self.venue)
        print("Seat Number:", self.seat_number)
        print("Price:", self.price)
        print("Reservation Status:", "Reserved" if self.is_reserved else "Not Reserved")
        
In this implementation, the "Ticket" class is defined with the specified attributes (ticket_id, event_name, event_date, venue, seat_number, price, is_reserved). The class includes three methods: reserve_ticket(), cancel_reservation(), and display_ticket_info(), which perform the desired functionalities.

Implementation of the "ShoppingCart" class:

# Python code

class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)

    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)

    def view_cart(self):
        for item in self.items:
            print(item)

    def clear_cart(self):
        self.items = []
        
In this implementation, the "ShoppingCart" class is defined with a single attribute items, which represents the list of items in the shopping cart. The class includes four methods: add_item(), remove_item(), view_cart(), and clear_cart(), which perform the desired functionalities.

Implementation of the "Student" class:

# Python code

class Student:
    def __init__(self, name, age, grade, student_id):
        self.name = name
        self.age = age
        self.grade = grade
        self.student_id = student_id
        self.attendance = {}

    def update_attendance(self, date, status):
        self.attendance[date] = status

    def get_attendance(self):
        return self.attendance

    def get_average_attendance(self):
        total_days = len(self.attendance)
        present_days = list(self.attendance.values()).count("Present")
        attendance_percentage = (present_days / total_days) * 100
        return attendance_percentage
        
In this implementation, the "Student" class is defined with the specified attributes (name, age, grade, student_id) and an additional attribute attendance to store the attendance record. The class includes three methods: update_attendance(), get_attendance(), and get_average_attendance(), which perform the desired functionalities.
